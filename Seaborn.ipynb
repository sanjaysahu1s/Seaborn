Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

#Answer

Seaborn is a powerful data visualization library in Python that is built on top of Matplotlib. It provides a high-level interface for creating informative and visually appealing statistical graphics. Here are five commonly used plots in Seaborn along with their uses:

1. **Line Plot**: A line plot displays the relationship between two continuous variables over a continuous interval. It is useful for visualizing trends or patterns in data over time or any continuous scale.
   - Uses: Trend analysis, time series data visualization, showing the change of a variable over time.

2. **Bar Plot**: A bar plot represents categorical data using rectangular bars. The height of each bar corresponds to the frequency or count of the categories being plotted.
   - Uses: Comparing the distribution of categorical variables, visualizing the frequency or count of different categories.

3. **Scatter Plot**: A scatter plot represents the relationship between two continuous variables by placing individual data points on a Cartesian plane. Each data point is represented by a marker, and the position on the plot represents the values of the two variables.
   - Uses: Identifying patterns or relationships between two continuous variables, visualizing correlations.

4. **Histogram**: A histogram is a graphical representation of the distribution of a continuous variable. It divides the range of values into bins and displays the count or frequency of data points falling into each bin.
   - Uses: Understanding the distribution of data, identifying skewness or outliers, visualizing data density.

5. **Heatmap**: A heatmap is a matrix-like plot that represents the relationship between two variables using color encoding. It is commonly used to visualize correlations or patterns in a dataset.
   - Uses: Displaying the correlation between multiple variables, identifying clusters or patterns in data.

These are just a few examples of the many plots available in Seaborn. Each plot serves a different purpose and can be customized further to suit specific needs.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. 

Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
#Answer

import seaborn as sns

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Plot line plot with different events and regions
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")

# Display the plot
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. 

Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

Note: pclass, age, and fare are columns in the titanic dataset.

```python
#Answer

import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Plot box plots for age and fare based on pclass
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Box plot for age
sns.boxplot(ax=axes[0], data=titanic_data, x="pclass", y="age")
axes[0].set_title("Box Plot of Age by Pclass")

# Box plot for fare
sns.boxplot(ax=axes[1], data=titanic_data, x="pclass", y="fare")
axes[1].set_title("Box Plot of Fare by Pclass")

# Adjust the layout
plt.tight_layout()

# Display the plot
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
#Answer

import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot histogram for 'price' column with hue='cut'
sns.histplot(data=diamonds_data, x="price", hue="cut", kde=True)

# Set plot title and labels
plt.title("Histogram of Price with Cut")
plt.xlabel("Price")
plt.ylabel("Count")

# Display the plot
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
#Answer

import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Reshape the data into a pivot table
flights_pivot = flights_data.pivot("month", "year", "passengers")

# Plot a heatmap
sns.heatmap(flights_pivot, cmap="YlGnBu")

# Set plot title and labels
plt.title("Number of Passengers by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")

# Display the plot
plt.show()
```
